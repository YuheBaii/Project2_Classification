# 1.2 Model Architecture Comparison

Compare different model architectures with Light Augmentation + AdamW

```python
import json
import matplotlib.pyplot as plt
import numpy as np

experiments = {
    'SimpleCNN': 'experiments/2025-10-24_13-09-40_dogs_vs_cats_simple_cnn_binary/history.json',
    'CNN_BN': 'experiments/2025-10-26_14-51-41_dogs_vs_cats_cnn_bn/history.json',
    'VGG-16': 'experiments/2025-11-01_10-38-32_dogs_vs_cats_vgg/history.json',
    'ResNet18': 'experiments/2025-11-01_12-36-25_dogs_vs_cats_resnet18/history.json',
    'ResNeXt50': 'experiments/2025-11-01_13-36-36_dogs_vs_cats_resnext50/history.json',
    'DenseNet_Small': 'experiments/2025-11-03_05-55-53_dogs_vs_cats_densenet/history.json',
    'DenseNet121': 'experiments/2025-11-03_06-15-48_dogs_vs_cats_densenet/history.json'
}

histories = {}
for name, path in experiments.items():
    with open(path, 'r') as f:
        histories[name] = json.load(f)

colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b', '#e377c2']
fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# Training Accuracy
for (name, history), color in zip(histories.items(), colors):
    epochs = range(1, len(history['accuracy']) + 1)
    axes[0, 0].plot(epochs, history['accuracy'], label=name, linewidth=2, color=color, alpha=0.8)
axes[0, 0].set_title('Training Accuracy Comparison', fontsize=14, fontweight='bold')
axes[0, 0].set_xlabel('Epochs')
axes[0, 0].set_ylabel('Accuracy')
axes[0, 0].legend(loc='lower right')
axes[0, 0].grid(True, alpha=0.3)

# Validation Accuracy
for (name, history), color in zip(histories.items(), colors):
    epochs = range(1, len(history['val_accuracy']) + 1)
    axes[0, 1].plot(epochs, history['val_accuracy'], label=name, linewidth=2, color=color, alpha=0.8)
axes[0, 1].set_title('Validation Accuracy Comparison', fontsize=14, fontweight='bold')
axes[0, 1].set_xlabel('Epochs')
axes[0, 1].set_ylabel('Accuracy')
axes[0, 1].legend(loc='lower right')
axes[0, 1].grid(True, alpha=0.3)

# Training Loss
for (name, history), color in zip(histories.items(), colors):
    epochs = range(1, len(history['loss']) + 1)
    axes[1, 0].plot(epochs, history['loss'], label=name, linewidth=2, color=color, alpha=0.8)
axes[1, 0].set_title('Training Loss Comparison', fontsize=14, fontweight='bold')
axes[1, 0].set_xlabel('Epochs')
axes[1, 0].set_ylabel('Loss')
axes[1, 0].legend(loc='upper right')
axes[1, 0].grid(True, alpha=0.3)

# Validation Loss
for (name, history), color in zip(histories.items(), colors):
    epochs = range(1, len(history['val_loss']) + 1)
    axes[1, 1].plot(epochs, history['val_loss'], label=name, linewidth=2, color=color, alpha=0.8)
axes[1, 1].set_title('Validation Loss Comparison', fontsize=14, fontweight='bold')
axes[1, 1].set_xlabel('Epochs')
axes[1, 1].set_ylabel('Loss')
axes[1, 1].legend(loc='upper right')
axes[1, 1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("="*80)
print("Model Architecture Comparison: Light Aug + AdamW")
print("="*80)

results = []
for name, history in histories.items():
    final_train_acc = history['accuracy'][-1]
    final_val_acc = history['val_accuracy'][-1]
    best_val_acc = max(history['val_accuracy'])
    best_epoch = history['val_accuracy'].index(best_val_acc) + 1
    final_train_loss = history['loss'][-1]
    final_val_loss = history['val_loss'][-1]
    overfitting = final_train_acc - final_val_acc
    
    results.append({
        'name': name,
        'final_val_acc': final_val_acc,
        'best_val_acc': best_val_acc,
        'best_epoch': best_epoch,
        'overfitting': overfitting
    })
    
    print(f"\n[{name}]")
    print(f"  Final Train Acc: {final_train_acc:.4f}")
    print(f"  Final Val Acc: {final_val_acc:.4f}")
    print(f"  Best Val Acc: {best_val_acc:.4f} (Epoch {best_epoch})")
    print(f"  Final Train Loss: {final_train_loss:.4f}")
    print(f"  Final Val Loss: {final_val_loss:.4f}")
    print(f"  Overfitting Gap: {overfitting:.4f}")

results_sorted = sorted(results, key=lambda x: x['best_val_acc'], reverse=True)

print("\n" + "="*80)
print("Ranking by Best Validation Accuracy:")
print("="*80)
for i, result in enumerate(results_sorted, 1):
    print(f"{i}. {result['name']}: {result['best_val_acc']:.4f} (Epoch {result['best_epoch']})")

print("\n" + "="*80)
print("Key Findings:")
print("="*80)
best_model = results_sorted[0]
print(f"• Best Model: {best_model['name']} with {best_model['best_val_acc']:.4f} validation accuracy")
print(f"• Least Overfitting: {min(results, key=lambda x: x['overfitting'])['name']} " +
      f"(gap: {min(results, key=lambda x: x['overfitting'])['overfitting']:.4f})")
print(f"• Most Overfitting: {max(results, key=lambda x: x['overfitting'])['name']} " +
      f"(gap: {max(results, key=lambda x: x['overfitting'])['overfitting']:.4f})")
print("="*80)
```

```python
import json
import matplotlib.pyplot as plt
import numpy as np

adamw_lr_experiments = {
    'AdamW (lr=0.001)': 'experiments/2025-11-01_17-26-41_dogs_vs_cats_cnn_bn/history.json',
    'AdamW (lr=0.01)': 'experiments/2025-11-01_16-00-00_dogs_vs_cats_cnn_bn/history.json',
    'AdamW (lr=0.1)': 'experiments/2025-11-01_16-49-38_dogs_vs_cats_cnn_bn/history.json'
}

adamw_histories = {}
for name, path in adamw_lr_experiments.items():
    with open(path, 'r') as f:
        adamw_histories[name] = json.load(f)

colors = ['#2ca02c', '#1f77b4', '#d62728']
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

# Training Accuracy
for (name, history), color in zip(adamw_histories.items(), colors):
    epochs = range(1, len(history['accuracy']) + 1)
    axes[0, 0].plot(epochs, history['accuracy'], label=name, linewidth=2.5, color=color, alpha=0.8)
axes[0, 0].set_title('Training Accuracy - AdamW Learning Rate', fontsize=13, fontweight='bold')
axes[0, 0].set_xlabel('Epochs')
axes[0, 0].set_ylabel('Accuracy')
axes[0, 0].legend(loc='lower right', fontsize=11)
axes[0, 0].grid(True, alpha=0.3)

# Validation Accuracy
for (name, history), color in zip(adamw_histories.items(), colors):
    epochs = range(1, len(history['val_accuracy']) + 1)
    axes[0, 1].plot(epochs, history['val_accuracy'], label=name, linewidth=2.5, color=color, alpha=0.8)
axes[0, 1].set_title('Validation Accuracy - AdamW Learning Rate', fontsize=13, fontweight='bold')
axes[0, 1].set_xlabel('Epochs')
axes[0, 1].set_ylabel('Accuracy')
axes[0, 1].legend(loc='lower right', fontsize=11)
axes[0, 1].grid(True, alpha=0.3)

# Training Loss
for (name, history), color in zip(adamw_histories.items(), colors):
    epochs = range(1, len(history['loss']) + 1)
    axes[1, 0].plot(epochs, history['loss'], label=name, linewidth=2.5, color=color, alpha=0.8)
axes[1, 0].set_title('Training Loss - AdamW Learning Rate', fontsize=13, fontweight='bold')
axes[1, 0].set_xlabel('Epochs')
axes[1, 0].set_ylabel('Loss')
axes[1, 0].legend(loc='upper right', fontsize=11)
axes[1, 0].grid(True, alpha=0.3)

# Validation Loss
for (name, history), color in zip(adamw_histories.items(), colors):
    epochs = range(1, len(history['val_loss']) + 1)
    axes[1, 1].plot(epochs, history['val_loss'], label=name, linewidth=2.5, color=color, alpha=0.8)
axes[1, 1].set_title('Validation Loss - AdamW Learning Rate', fontsize=13, fontweight='bold')
axes[1, 1].set_xlabel('Epochs')
axes[1, 1].set_ylabel('Loss')
axes[1, 1].legend(loc='upper right', fontsize=11)
axes[1, 1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("="*70)
print("Learning Rate Comparison: CNN_BN + Light Aug + AdamW")
print("="*70)

adamw_results = []
for name, history in adamw_histories.items():
    final_train_acc = history['accuracy'][-1]
    final_val_acc = history['val_accuracy'][-1]
    best_val_acc = max(history['val_accuracy'])
    best_epoch = history['val_accuracy'].index(best_val_acc) + 1
    final_train_loss = history['loss'][-1]
    final_val_loss = history['val_loss'][-1]
    
    adamw_results.append({
        'name': name,
        'best_val_acc': best_val_acc,
        'best_epoch': best_epoch
    })
    
    print(f"\n[{name}]")
    print(f"  Final Train Acc: {final_train_acc:.4f}")
    print(f"  Final Val Acc: {final_val_acc:.4f}")
    print(f"  Best Val Acc: {best_val_acc:.4f} (Epoch {best_epoch})")
    print(f"  Final Train Loss: {final_train_loss:.4f}")
    print(f"  Final Val Loss: {final_val_loss:.4f}")
    print(f"  Overfitting Gap: {final_train_acc - final_val_acc:.4f}")

adamw_results_sorted = sorted(adamw_results, key=lambda x: x['best_val_acc'], reverse=True)

print("\n" + "="*70)
print("Learning Rate Ranking:")
print("="*70)
for i, result in enumerate(adamw_results_sorted, 1):
    print(f"{i}. {result['name']}: {result['best_val_acc']:.4f} (Epoch {result['best_epoch']})")

print("\n" + "="*70)
print("Key Findings:")
print("="*70)
best_lr = adamw_results_sorted[0]
print(f"• Best Learning Rate: {best_lr['name']}")
print(f"• Performance: {best_lr['best_val_acc']:.4f} at Epoch {best_lr['best_epoch']}")
print("="*70)
```

## 1.3.4 Overall Comparison: All Configurations

```python
import json
import matplotlib.pyplot as plt
import numpy as np

all_experiments = {
    'AdamW (lr=0.001)': 'experiments/2025-11-01_17-26-41_dogs_vs_cats_cnn_bn/history.json',
    'AdamW (lr=0.01)': 'experiments/2025-11-01_16-00-00_dogs_vs_cats_cnn_bn/history.json',
    'AdamW (lr=0.1)': 'experiments/2025-11-01_16-49-38_dogs_vs_cats_cnn_bn/history.json',
    'SGD (lr=0.001)': 'experiments/2025-11-01_17-47-32_dogs_vs_cats_cnn_bn/history.json',
    'SGD (lr=0.01)': 'experiments/2025-11-02_04-45-10_dogs_vs_cats_cnn_bn/history.json'
}

all_histories = {}
for name, path in all_experiments.items():
    with open(path, 'r') as f:
        all_histories[name] = json.load(f)

colors = ['#2ca02c', '#1f77b4', '#d62728', '#9467bd', '#ff7f0e']
fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# Training Accuracy
for (name, history), color in zip(all_histories.items(), colors):
    epochs = range(1, len(history['accuracy']) + 1)
    axes[0, 0].plot(epochs, history['accuracy'], label=name, linewidth=2, color=color, alpha=0.8)
axes[0, 0].set_title('Training Accuracy - All Configurations', fontsize=14, fontweight='bold')
axes[0, 0].set_xlabel('Epochs')
axes[0, 0].set_ylabel('Accuracy')
axes[0, 0].legend(loc='lower right', fontsize=10)
axes[0, 0].grid(True, alpha=0.3)

# Validation Accuracy
for (name, history), color in zip(all_histories.items(), colors):
    epochs = range(1, len(history['val_accuracy']) + 1)
    axes[0, 1].plot(epochs, history['val_accuracy'], label=name, linewidth=2, color=color, alpha=0.8)
axes[0, 1].set_title('Validation Accuracy - All Configurations', fontsize=14, fontweight='bold')
axes[0, 1].set_xlabel('Epochs')
axes[0, 1].set_ylabel('Accuracy')
axes[0, 1].legend(loc='lower right', fontsize=10)
axes[0, 1].grid(True, alpha=0.3)

# Training Loss
for (name, history), color in zip(all_histories.items(), colors):
    epochs = range(1, len(history['loss']) + 1)
    axes[1, 0].plot(epochs, history['loss'], label=name, linewidth=2, color=color, alpha=0.8)
axes[1, 0].set_title('Training Loss - All Configurations', fontsize=14, fontweight='bold')
axes[1, 0].set_xlabel('Epochs')
axes[1, 0].set_ylabel('Loss')
axes[1, 0].legend(loc='upper right', fontsize=10)
axes[1, 0].grid(True, alpha=0.3)

# Validation Loss
for (name, history), color in zip(all_histories.items(), colors):
    epochs = range(1, len(history['val_loss']) + 1)
    axes[1, 1].plot(epochs, history['val_loss'], label=name, linewidth=2, color=color, alpha=0.8)
axes[1, 1].set_title('Validation Loss - All Configurations', fontsize=14, fontweight='bold')
axes[1, 1].set_xlabel('Epochs')
axes[1, 1].set_ylabel('Loss')
axes[1, 1].legend(loc='upper right', fontsize=10)
axes[1, 1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("="*80)
print("Complete Optimizer & Learning Rate Comparison: CNN_BN + Light Aug")
print("="*80)

all_results = []
for name, history in all_histories.items():
    final_train_acc = history['accuracy'][-1]
    final_val_acc = history['val_accuracy'][-1]
    best_val_acc = max(history['val_accuracy'])
    best_epoch = history['val_accuracy'].index(best_val_acc) + 1
    final_train_loss = history['loss'][-1]
    final_val_loss = history['val_loss'][-1]
    overfitting = final_train_acc - final_val_acc
    
    all_results.append({
        'name': name,
        'final_val_acc': final_val_acc,
        'best_val_acc': best_val_acc,
        'best_epoch': best_epoch,
        'overfitting': overfitting,
        'convergence_speed': best_epoch
    })
    
    print(f"\n[{name}]")
    print(f"  Final Train Acc: {final_train_acc:.4f}")
    print(f"  Final Val Acc: {final_val_acc:.4f}")
    print(f"  Best Val Acc: {best_val_acc:.4f} (Epoch {best_epoch})")
    print(f"  Final Train Loss: {final_train_loss:.4f}")
    print(f"  Final Val Loss: {final_val_loss:.4f}")
    print(f"  Overfitting Gap: {overfitting:.4f}")

all_results_sorted = sorted(all_results, key=lambda x: x['best_val_acc'], reverse=True)

print("\n" + "="*80)
print("Overall Ranking by Best Validation Accuracy:")
print("="*80)
for i, result in enumerate(all_results_sorted, 1):
    print(f"{i}. {result['name']}: {result['best_val_acc']:.4f} (Epoch {result['best_epoch']})")

print("\n" + "="*80)
print("Key Findings:")
print("="*80)
best_config = all_results_sorted[0]
fastest_convergence = min(all_results, key=lambda x: x['convergence_speed'])
least_overfit = min(all_results, key=lambda x: x['overfitting'])

print(f"• Best Configuration: {best_config['name']}")
print(f"  - Best Val Acc: {best_config['best_val_acc']:.4f} at Epoch {best_config['best_epoch']}")
print(f"\n• Fastest Convergence: {fastest_convergence['name']}")
print(f"  - Reached best at Epoch {fastest_convergence['convergence_speed']}")
print(f"\n• Least Overfitting: {least_overfit['name']}")
print(f"  - Overfitting Gap: {least_overfit['overfitting']:.4f}")

print("\n" + "="*80)
print("Summary Insights:")
print("="*80)
print("• AdamW generally achieves better validation accuracy than SGD")
print("• Learning rate 0.01 provides the best balance for both optimizers")
print("• Too high LR (0.1) causes training instability")
print("• Too low LR (0.001) leads to slower convergence")
print("="*80)
```
